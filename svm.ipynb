```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.metrics import confusion_matrix, classification_report

%pylab inline 
%matplotlib inline
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
df.head()
```

```python
df["target"].value_counts()
```

```python
sns.FacetGrid(df, hue="target",height=6).map(plt.scatter, "petal length (cm)", "petal width (cm)").add_legend()
```

```python
from sklearn import svm
model = svm.SVC(kernel='linear', C=0.1)
model.fit(iris.data, iris.target)
```

```python
print(model.score(iris.data, iris.target))
```

```python
predicted = model.predict(iris.data)
expected = iris.target
```

```python
print(classification_report(expected, predicted))
print(confusion_matrix(expected, predicted))
```

#### Parameter Tuning

```python
X = iris.data[:, :2] # taking only the first two features
y = iris.target
```

```python
def MySVMClassifier(my_kernel, my_C, my_gamma):
    svc = svm.SVC(kernel=my_kernel, C=my_C,gamma=my_gamma)
    svc.fit(X, y)
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
 
    plt.subplot(1, 1, 1)
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.show() 
```

```python
from ipywidgets import interact
i = interact(MySVMClassifier, my_kernel=['linear','rbf'],my_C=(0.001,10), my_gamma=(1,100))
```
